```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# --- Configura tus datos aquí ---
datos = [10.5, 11.2, 9.8, 10.1, 11.5, 10.9, 10.3]
media_hipotetica = 10.0
# Si conoces la desviación estándar poblacional, descomenta la siguiente línea y asigna un valor.
# Si la dejas comentada, se realizará una prueba t.
# desviacion_estandar_poblacional = 0.8

print("--- Prueba de Medias ---")
n = len(datos)
media_muestra = np.mean(datos)

print(f"Número de datos (n): {n}")
print(f"Media de la muestra (x̄): {media_muestra:.4f}")
print(f"Media hipotética (μ₀): {media_hipotetica}")

# Determinar si es una prueba Z o t
try:
    sigma = desviacion_estandar_poblacional
    print("\nRealizando Prueba Z (desviación estándar poblacional conocida)...")
    z_score = (media_muestra - media_hipotetica) / (sigma / np.sqrt(n))
    valor_p = 2 * (1 - norm.cdf(abs(z_score)))
    print(f"Estadístico Z: {z_score:.4f}")
    print(f"Valor p: {valor_p:.4f}")

    # Gráfico de la distribución normal
    fig, ax = plt.subplots(figsize=(8, 5))
    x = np.linspace(-4, 4, 100)
    ax.plot(x, norm.pdf(x, 0, 1), label='Distribución Normal Estándar')
    ax.axvline(z_score, color='red', linestyle='--', label=f'Estadístico Z: {z_score:.4f}')
    ax.set_title('Distribución Normal Estándar (Prueba Z)')
    ax.set_xlabel('Valor Z')
    ax.set_ylabel('Densidad de Probabilidad')
    ax.legend()
    plt.grid(True)
    plt.show()

except NameError:
    print("\nRealizando Prueba t (desviación estándar poblacional desconocida)...")
    s_muestra = np.std(datos, ddof=1)
    t_statistic = (media_muestra - media_hipotetica) / (s_muestra / np.sqrt(n))
    grados_libertad = n - 1
    print(f"Estadístico t: {t_statistic:.4f}")
    print(f"Grados de libertad: {grados_libertad}")

    # Gráfico de la distribución t de Student
    fig, ax = plt.subplots(figsize=(8, 5))
    x = np.linspace(-4, 4, 100)
    ax.plot(x, t.pdf(x, df=grados_libertad), label=f'Distribución t (df={grados_libertad})')
    ax.axvline(t_statistic, color='red', linestyle='--', label=f'Estadístico t: {t_statistic:.4f}')
    ax.set_title('Distribución t de Student (Prueba t)')
    ax.set_xlabel('Valor t')
    ax.set_ylabel('Densidad de Probabilidad')
    ax.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# --- Configura tus datos aquí ---
datos = [5.1, 4.9, 5.5, 4.8, 5.0, 5.3, 5.2]
varianza_hipotetica = 0.1

print("--- Prueba de Varianza ---")
n = len(datos)
varianza_muestra = np.var(datos, ddof=1)
grados_libertad = n - 1
chi2_statistic = (n - 1) * varianza_muestra / varianza_hipotetica

print(f"Número de datos (n): {n}")
print(f"Varianza de la muestra (s²): {varianza_muestra:.4f}")
print(f"Varianza hipotética (σ₀²): {varianza_hipotetica}")
print(f"Grados de libertad: {grados_libertad}")
print(f"Estadístico Chi-cuadrado (χ²): {chi2_statistic:.4f}")

# Gráfico de la distribución Chi-cuadrado
fig, ax = plt.subplots(figsize=(8, 5))
x = np.linspace(0, chi2.ppf(0.99, grados_libertad) * 1.5, 100)
ax.plot(x, chi2.pdf(x, grados_libertad), 'r-', lw=2, label='Distribución Chi-cuadrado')
ax.axvline(chi2_statistic, color='blue', linestyle='--', label=f'Estadístico χ²: {chi2_statistic:.4f}')
ax.set_title('Prueba de Varianza Chi-cuadrado')
ax.set_xlabel('Valor')
ax.set_ylabel('Densidad de Probabilidad')
ax.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kstest

# --- Configura tus datos aquí (números entre 0 y 1) ---
datos = [0.12, 0.85, 0.33, 0.61, 0.44, 0.79, 0.92, 0.08, 0.50, 0.27]

print("--- Prueba de Uniformidad (Kolmogorov-Smirnov) ---")
estadistico_ks, valor_p = kstest(datos, 'uniform')

print(f"Estadístico de la prueba K-S: {estadistico_ks:.4f}")
print(f"Valor p: {valor_p:.4f}")
print("\nInterpretación:")
if valor_p < 0.05:
    print("El valor p es menor que 0.05. Los datos NO parecen seguir una distribución uniforme.")
else:
    print("El valor p es mayor que 0.05. Los datos parecen seguir una distribución uniforme.")

# Gráfico de la distribución empírica vs. teórica
n = len(datos)
datos_ordenados = np.sort(datos)
cdf_empirica = np.arange(1, n + 1) / n

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(datos_ordenados, cdf_empirica, label='CDF Empírica', marker='o')
ax.plot(np.linspace(0, 1, 100), np.linspace(0, 1, 100), label='CDF Uniforme Teórica', linestyle='--')
ax.set_title('Prueba de Uniformidad (Kolmogorov-Smirnov)')
ax.set_xlabel('Valor')
ax.set_ylabel('Probabilidad Acumulada')
ax.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# --- Configura tus parámetros aquí ---
semilla_cm = 2015
n_numeros_cm = 20

print("--- Algoritmo de Cuadrados Medios ---")
resultados_cm = []
semilla_str = str(semilla_cm).zfill(4)

for i in range(n_numeros_cm):
    cuadrado = int(semilla_str) ** 2
    cuadrado_str = str(cuadrado).zfill(8)
    medio = cuadrado_str[2:6]
    numeros_aleatorios = int(medio) / 10000
    resultados_cm.append(numeros_aleatorios)
    semilla_str = medio
    print(f"Iteración {i+1}: Semilla {semilla_str} -> Número: {numeros_aleatorios:.4f}")

# Gráfico de histograma
fig, ax = plt.subplots(figsize=(8, 5))
ax.hist(resultados_cm, bins=10, edgecolor='black', alpha=0.7)
ax.set_title('Distribución de Números de Cuadrados Medios')
ax.set_xlabel('Valor')
ax.set_ylabel('Frecuencia')
ax.grid(axis='y', alpha=0.75)
plt.show()
```
